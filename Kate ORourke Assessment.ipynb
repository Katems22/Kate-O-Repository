```python
import pandas as pd
```

```python
const = pd.read_csv("constructors.csv")
```

```python
pd.read_csv("constructors.csv")
```

```python
const_standings = pd.read_csv("constructor_standings.csv",index_col=-5)
```

```python
const_standings
```

```python
const_with_standings = const.merge(const_standings, on = "constructorId", suffixes = ("_const","_const_st"), how = "left")
```

```python
const_with_standings
```

```python
wins = const_with_standings[const_with_standings["position"] < 2.0]
```

```python
wins
```

```python
wins_summary = wins.groupby(["name"]).agg(total_wins=("position","sum"))
```

```python
wins_summary
```

```python
wins_summary.shape
```

```python
wins_in_order=wins_summary.sort_values("total_wins",ascending = False)
```

```python
wins_in_order
```
